```python
import pandas as pd
from pyhigh import get_elevation
import json

def df_to_json(cur_df, siteName):
    json_output = json.dumps(cur_df.to_dict(orient='records'))
    json_file = f'neko-app/public/data/{siteName}.json'
    with open(json_file, 'w') as f:
        f.write(json_output)

for site in ["pt","sd","gm"]:
    curDF = pd.read_excel("neko-app/public/data/nekor_data.xlsx", sheet_name=site)
    curDF["imgPath"] = curDF.apply(lambda r: f"./img/{site}/{site}_{str(r.name)}.png",axis=1)
    curDF["lat"] = curDF.apply(lambda r: r.coord.split(",")[0],axis=1)
    curDF["long"] = curDF.apply(lambda r: r.coord.split(",")[1],axis=1)
    curDF = curDF[['title', 'description', 'markerType', 'lat', 'long', 'imgPath']]
    df_to_json(curDF, site)
    display(curDF)
```
